```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('diabetes_model.csv') # accessing csv file
```

```python
data1.columns # checking the coloumns present
```

```python
data1.head() # checking first 5 rows
```

# Q1 Countplot

```python
sns.countplot(x='Outcome', data=data1)
```

# Q2. Box plot

```python
sns.boxplot(x='Outcome',y='Age',data=data1)
```

# Q3. Check for chances of getting diabetes using Diabetes Pedigree Function

```python
sns.swarmplot(y=data1['DiabetesPedigreeFunction'],x=data1['Outcome'],data=data1,s=3)
```

# Q4. Relationship with Glucose & Insulin with Outcome

```python
sns.scatterplot(x=data1['Insulin'],y=data1['Glucose'],hue=data1['Outcome'])
```

# Q5.Correlation of Features in Data

```python
sns.heatmap(data1.corr(),annot=True)
```
